```python
### Creando formato de tiempo a la columna fecha

df['month'] = pd.to_datetime(df['FECHA'])
df['month']
```

```python
### Agrupando por fecha (mes) y por pais para encontrar un promedio anuaL del Valor CIF. en Dolares.

df_t_avg = df.groupby(['Pais Fraccion',
                     pd.Grouper(key='month', freq='1m')
                      ]).agg({'Valor Cif. $.':[np.mean, np.median]})
df_t_avg
```
